# **Árbol de decisión**

En esta libreta se entrena un árbol de decisión en el conjunto de datos Iris. Mas información:
* https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html
* https://archive.ics.uci.edu/dataset/53/iris



1. Importar librerías.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import numpy as np
```

2. Cargar datos

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
iris.target_names
```

3. Preprocesamiento

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

4. Selección y entrenamiento del modelo

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
# Exactitud del modelo en el entrenamiento.
train_accuracy = clf.score(X_train, y_train)
print(train_accuracy)
```

5. Prueba del modelo.

```python
y_pred = clf.predict(X_test)
```

```python
# Obtener la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
# Mostrar la matriz de confusión de resultados de la prueba
plt.figure(figsize=(5,4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['setosa', 'versicolor', 'virginica'], yticklabels=['setosa', 'versicolor', 'virginica'])
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.title("Matriz de Confusión (Entrenamiento)")
plt.show()
```

```python
#Predicción 1
print(y_pred[1])
print(y_test[1])
```

```python
#Predicción 2
print(y_pred[20])
print(y_test[20])
```

```python
#Predicción 3
print(y_pred[-1])
print(y_test[-1])
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=iris.target_names))
```

Visualización del árbol de decisión.

```python
plt.figure(figsize=(20, 15))  # Increase figure size for better resolution
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree for Iris Species Prediction")
plt.savefig("decision_tree.png", dpi=300)  # Increase DPI for better resolution
plt.show()
```
