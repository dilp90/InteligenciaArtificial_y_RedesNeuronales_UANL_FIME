# **Tutorial para librería Matplotlib**

Matplotlib es una biblioteca integral para crear visualizaciones estáticas, animadas e interactivas en Python.

**Ejercicio:**

Visualización de datos y resultados con Matplotlib en Python para Inteligencia Artificial.

**Objetivo:**

Familiarizarte con el uso de Matplotlib para la visualización de datos y resultados en proyectos de inteligencia artificial. Aprenderás a crear gráficos básicos y avanzados que son fundamentales para analizar el comportamiento de modelos de IA.

**1. Importar librerías necesarias**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

**2. Gráficos básicos:**

*Gráfico de líneas*: Crea un gráfico de líneas para visualizar una función matemática, como la función seno.

```python
x = np.linspace(0, 10, 100)  # Genera 100 puntos entre 0 y 10
y = np.sin(x)  # Calcula el seno de cada punto

plt.plot(x, y, label="Seno de x")  # Dibuja el gráfico
plt.title("Gráfico de Seno")  # Título
plt.xlabel("x")  # Etiqueta del eje x
plt.ylabel("sin(x)")  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda
plt.grid(True)  # Muestra la cuadrícula
plt.show()  # Muestra el gráfico
```

**3. Gráficos de dispersión:**

*Gráfico de dispersión (Scatter plot)*: Crea un gráfico de dispersión para visualizar cómo se distribuyen los puntos en dos dimensiones.

```python
x = np.random.rand(100)  # Genera 100 puntos aleatorios en x
y = np.random.rand(100)  # Genera 100 puntos aleatorios en y

plt.scatter(x, y, color='blue', label="Datos Aleatorios")  # Dibuja el gráfico de dispersión
plt.title("Gráfico de Dispersión")  # Título
plt.xlabel("x")  # Etiqueta del eje x
plt.ylabel("y")  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda
plt.grid(True)  # Muestra la cuadrícula
plt.show()  # Muestra el gráfico
```

**4. Gráficos de barras:**

*Gráfico de barras*: Visualiza un conjunto de datos de barras, útil para comparar categorías.

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [5, 7, 3, 9, 6]

plt.bar(categorias, valores, color='orange')  # Dibuja el gráfico de barras
plt.title("Gráfico de Barras")  # Título
plt.xlabel("Categorías")  # Etiqueta del eje x
plt.ylabel("Valores")  # Etiqueta del eje y
plt.show()  # Muestra el gráfico
```

**5. Histograma:**

*Histograma*: Utiliza un histograma para ver la distribución de un conjunto de datos, útil para observar la distribución de características en un modelo de IA.

```python
datos = np.random.randn(1000)  # Genera 1000 puntos con distribución normal

plt.hist(datos, bins=30, color='green', alpha=0.7)  # Dibuja el histograma
plt.title("Histograma de Datos Aleatorios")  # Título
plt.xlabel("Valor")  # Etiqueta del eje x
plt.ylabel("Frecuencia")  # Etiqueta del eje y
plt.show()  # Muestra el gráfico
```

**6. Subgráficos (Subplots):**

*Subgráficos (Multiple Plots)*: Muestra varios gráficos en la misma ventana para comparar diferentes visualizaciones.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))  # Crea una fila con dos gráficos

axs[0].plot(x, y1, label="Seno de x")
axs[0].set_title("Gráfico de Seno")
axs[0].legend()

axs[1].plot(x, y2, label="Coseno de x", color='red')
axs[1].set_title("Gráfico de Coseno")
axs[1].legend()

plt.tight_layout()  # Ajusta el espaciado
plt.show()  # Muestra los gráficos
```

**7. Gráfico de regresión (si tienes un modelo de IA)**:

*Gráfico de regresión*: Visualiza los resultados de una regresión lineal para evaluar el desempeño de un modelo de IA.

```python
# Datos de ejemplo para regresión
X = np.random.rand(100, 1) * 10  # 100 puntos aleatorios entre 0 y 10
y = 2 * X + 1 + np.random.randn(100, 1)  # Relación lineal con algo de ruido

model = LinearRegression()
model.fit(X, y)  # Ajusta el modelo de regresión

y_pred = model.predict(X)  # Predicciones del modelo

plt.scatter(X, y, color='blue', label="Datos reales")  # Muestra los datos reales
plt.plot(X, y_pred, color='red', label="Regresión lineal")  # Muestra la línea de regresión
plt.title("Regresión Lineal")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
